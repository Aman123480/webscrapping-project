```python
#pages of url in car24

for i in range(1,31):
    URL='https://www.cars24.com/buy-used-car/?sort=P&search=HONDA&storeCityId=2&pinId=110001&page={}'.format(i)
    print(URL)
```

```python
Title = []
Emi_month=[]
Feature=[]
brand=[]
Overall_Amt=[]

for i in range (1,31):
    url='https://www.cars24.com/buy-used-car/?sort=P&search=HONDA&storeCityId=2&pinId=110001&page={}'.format(i)
    pages = requests.get(url)
    soup = BeautifulSoup(pages.text)

    for x in soup.find_all('div',class_="_1l4fi"):

        Titles=x.find("h2",class_="_3FpCg")
        if Titles is None:
            Title.append(np.NaN)
        else:
            Title.append(Titles.text)
            
        Emi_months=x.find("div",class_="_2HFRN")
        if Emi_months is None:
            Emi_month.append(np.NaN)
        else:
            Emi_month.append(Emi_months.text)
                
            
        Features=x.find("ul",class_="bVR0c")
        if Features is None:
            Feature.append(np.NaN)
        else:
            Feature.append(Features.text)
                
                
        Overall_Amts=x.find("div",class_="_7udZZ")
        if Overall_Amts is None:
            Overall_Amt.append(np.NaN)
        else:
            Overall_Amt.append(Overall_Amts.text)
             
```

```python
df = pd.DataFrame({'Title':Title,'Emi_month':Emi_month,'Feature':Feature,'Overall_Amt':Overall_Amt})
```

```python
Overall_Amts.text
```

```python

df['Year of Manufacture'] = df['Title'].apply(lambda x: x.split(' ')[0])
df['Brand'] = df['Title'].apply(lambda x: x.split(' ')[1])
df['Model'] = df['Title'].apply(lambda x: x.split(' ')[2])
df['Number of Km'] = df['Feature'].apply(lambda x: x.split(' ')[0])
df['owner'] = df['Feature'].apply(lambda x: x.split(' ')[1])
```

```python
#load a dataframe from csv

df=pd.read_csv('car24.csv')
```

```python
df.head()
```

```python
regex = 'Petrol|Diesel(?:)'
df['Fuel type'] = df['Feature'].apply(lambda x: re.findall(regex,x))
df['Fuel type'] = df['Fuel type'].apply(lambda x:''.join(x))
```

```python
df['Fuel type']
```

```python
df.head()
```

```python
df['Emi_month'] = df['Emi_month'].apply(lambda x: x.replace(',',''))
```

```python
df['Emi_month']
```

```python
df.head()
```

```python
df['Emi_month'] = df['Emi_month'].apply(lambda x: x.replace('₹',''))
df['Emi_month'] = df['Emi_month'].apply(lambda x: x.replace('month',''))
df['Emi_month'] = df['Emi_month'].apply(lambda x: x.replace('/',''))
```

```python
df.head()
```

```python
df['owner'] = df['owner'].apply(lambda x: x.replace('km',''))
```

```python
df.head()
```

```python
df['Overall_Amt'] = df['Overall_Amt'].apply(lambda x: x.replace('₹',''))
df['Overall_Amt'] = df['Overall_Amt'].apply(lambda x: x.replace(',',''))
```

```python
df.drop(['Title','Feature'], axis = 1)
```

```python
df['Number of Km'] = df['Number of Km'].apply(lambda x: x.replace(',',''))
```

```python
df['Number of Km'] = df['Number of Km'].astype(float)
df['Overall_Amt'] = df['Overall_Amt'].astype(float)
df['Emi_month'] = df['Emi_month'].astype(float)
df.head()
```
